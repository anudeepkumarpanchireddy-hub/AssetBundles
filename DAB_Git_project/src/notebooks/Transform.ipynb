```python
df1 = spark.read.table("dev.default.rider_details")
df1.display()
```

```python
df2 = spark.read.table("dev.default.vehicle_details")
df2.display()
```

```python
result_df = df2.join(df1, on="ride_id", how="inner")
result_df.display()
result_df.write.mode("overwrite").saveAsTable("dev.default.rider_vehicle_details")
```
